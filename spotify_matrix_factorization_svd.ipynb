```python
frames = [df, behavior, mul_label]
```

```python
result = pd.concat(frames,axis = 1)
```

```python
result['rating'] = result['skip_score'] + result['liking']
```

```python
len(result)
```

```python
result2 = result.drop_duplicates('session_id')
```

```python
result2
```

```python
R_df = result2.pivot(index = 'session_id', columns ='track_id', values = 'rating').fillna(0)
R_df
```

```python
R = R_df.as_matrix()
user_ratings_mean = np.mean(R, axis = 1)
R_demeaned = R - user_ratings_mean.reshape(-1, 1)
```

```python
R_demeaned
```

```python
from scipy.sparse.linalg import svds
U, sigma, Vt = svds(R_demeaned, k = 50)
```

```python
pd.DataFrame(U)
```

```python
pd.DataFrame(sigma)
```

```python
pd.DataFrame(Vt)
```

```python
sigma = np.diag(sigma)
```

```python
sigma
```

```python
all_user_predicted_ratings = np.dot(np.dot(U, sigma), Vt) + user_ratings_mean.reshape(-1, 1)
preds_df = pd.DataFrame(all_user_predicted_ratings, columns = R_df.columns)
```
